## Introduction

Suppose one wishes to compute the values $f(x)$ for a large number of arguments $x$, but that either:
- The function $f$ is not known explicitly (e.g. $f(t)$ is the average tempeature on earth at a time $t$...)
- It takes a long time to evaluate $f$ (for example when it is the output of a complicated code)

However, assume we already know the values $y_k = f(x_k)$ for $x_0, \cdots, x_N$. The arguments $(x_k)$ are called the **data points**. In practice, those values may come from experimental measures or previous runs of the code. 

In the interpolation method, we will approximate $f(x)$ for any $x$ by

$$f(x) \approx P(x)$$

where $P$ is a function that **exactly agrees** with $f$ at every data point $(x_k)$, that is, satisfying $P(x_k) = f(x_k)$ for all $0 \leq k \leq N$. 

- If $P$ is chosen as the polynomial of minimal degree satisfying this condition, this is the Lagrange interpolation method. 
- If $P$ is chosen such its restriction to each interval $]x_k,x_{k+1}[$ is a polynomial, this is piecewise polynomial interpolation. 

In the least-square method, we will approximate $f(x)$ for any $x$ by a model of the form 

$$f(x) \approx \beta_0 + \beta_1 x + \cdots + \beta_N x^N $$

For example affine regression is a least-square method whith two parameters $\beta_0$ and $\beta_1$, and with

$$f(x) \approx \beta_0 + \beta_1 x.$$

Contrary to the interpolation approach, one does not enforce exact agreement between the model polynomial and  $f$ at $x_k$ (because in general we use a polynomial of small degree). However, the values of the parameters $\beta_i$ are chosen so as to maximize this agreement (or, equivalently, to minimize the error). That is, we use the known data $(x_k)$ and $(y_k)$ to "tune" the parameters $(\beta_i)$ to be used later for new  unknown values $x$.

## Lagrange interpolation

### Introduction

In this first section, we choose to approximate the function $f$ by a polynomial $P$ on the whole interval, that coincides with $f$ at given *interpolation points*. We will address questions such as: how to choose the degree of this polynomial ? Is it unique ? How can we efficiently compute its coefficients from the knowledge of $(y_k)$ and $(x_k)$ ? Under which conditions is the error $\lvert P(x) - f(x) \rvert$ sufficiently small ? If we are free to choose the points $(x_k)$, is there a particularly good choice ?

### Interpolation polynomial of minimal degree

#### Existence and unicity of the Lagrange polynomial

<div  markdown=1 class="Thm">  Let $x_0, x_1, \cdots, x_n \in [a,b]$ a set of **interpolation points** and $y_0,y_1,\cdots,y_n$ a set of arbitrary **real numbers**. There exists a unique real polynomial $p_n$ of degree at most $n$ such that, $p_n(x_i) = y_i$ for all $0 \leq i \leq n.$ This polynomial is explicitly given by 

$$p_n(X) = \sum_{i=0}^n y_i L_i(X)$$

where $(L_i)_i$ are the elementary *Lagrange interpolation* polynomial given by 

$$ L_i(X) = \prod_{j \neq i}\frac{X - x_j}{x_i - x_j}, \quad 0 \leq i \leq n$$

#### Error bound

<div  markdown=1 class="Thm"> De Rolle

<div  markdown=1 class="Thm"> De Rolle généralisé (peut-être à enlever donc)

<div  markdown=1 class="Thm">  Let $f : [a,b] \to \mathbb{R}$ be $n+1$ times differentiable and let $P_n$ the Lagrange polynomial associated with the points $(x_i, f(x_i))_{0\leq i \leq n}$. Then, for every $y$ in $[a,b]$, there exists $\xi \in ]a,b[$ such that 

$$f(y) - p(y) = (y-x_0)(y-x_1)\cdots(y-x_n) \frac{f^{(n+1)}(\xi)}{(n+1)!}$$

<div  markdown=1 class="Thm">  Using the same notations as in the previous theorem, we have, for all $y \in [a,b]$, 

$$\lvert f(y) - p(y) \rvert \leq \frac{(b-a)^n}{n!} \sup_{[a,b]} \lvert f^{(n+1)} \rvert $$

### Computation of the Lagrange polynomial

#### Preliminaries : Hörner algorithm

In this section, we will need to evaluate polynomials of the form 

$$P(X) = a_0 + a_1 X + \cdots + a_N X^N$$

with given coefficients $(a_0,\cdots,a_N)$. For this, we rewrite 

$$P(X) = a_0 + X(a_1 + X ( a_2 + X ( \cdots ))))$$

Leading to the Hörner algorithm for evaluating the polynomial

<div  markdown=1 class="Algo">
**Hörner algorithm.** Evaluates  a polynomial with coefficients $(a_k)_k$ at $X$.


\begin{align}
INPUT:&\quad (a_k)_{0\leq k \leq N-1}, X\\
INIT:
&\quad R = a_{N-1}\\
FOR: &\quad k = 1 .. N-1 \\
&\quad\quad R = R X + a_{N-1-k}\\
RETURN:&\quad R\\
\end{align}

<div markdown=1 class="DoIt"> 1°) Write in the next cell a function with syntax `R = eval_Horner(a,X)` implementing the previous algorithm. 

```python
## Answer
def eval_Horner(a,X):
    N = a.size
    R = a[-1]
    for k in range(1,N):
        R = R*X + a[-1 -k]
    return R
```

Test your function by evaluating the following cell. 

```python
a = np.array([1,2,3]) # P(X) = 1 + 2X + 3X^2
x = np.array([0, 1, 2, 3, 6]) 
solution = np.array([1,6,17,34,121])
R = eval_Horner(a,x)
print('your code output :', R)
print('The solution :', solution)
if np.allclose(solution,R):
    print(':-)')
else:
    print(':-()')
```

#### The Van der Monde method

Assume we have $N = 21$ equispaced data points in the interval $[0,1]$, that is $x_k = \frac{k}{20}$ for $0 \leq k \leq N-1$. 

```python
a = -1; b= 1; N = 21; x_interp = np.linspace(a,b,N);
print(x_interp)
```

Let us choose $(y_i)_i$ as the values of some function $f$ at $(x_i)_i$. Here we take $f(x) = sin(2\pi x)$.

```python
def sineWave(x):
    return np.sin(2*pi*x)

y = sineWave(x_interp)
plt.figure(figsize=(10,5))
plt.plot(x_interp,y,'+')
plt.title('Data to interpolate')
plt.xlabel('Interpolation points')
plt.xticks(x_interp)
plt.ylabel('Interpolation values')
plt.show()
```

We now wish to compute the polynomial with minimal degree passing through the points on the previous graph using the Van der Monde method. 

<div markdown=1 class="DoIt"> 2°) Define in the next cell a function with syntax `M = VdM(x)`, taking in argument an array of real values and computing the vandermonde matrix 

$$\begin{pmatrix} 1&1&\cdots & 1 &1\\x_0&x_1&\cdots &x_{n-1}&x_{n}\\\vdots&\vdots&\ddots&\vdots&\vdots\\x_0^{n-1}&x_1^{n-1}&\cdots&x_{n-1}^{n-1}& x_n^{n-1}\\x_0^{n}&x_1^{n}&\cdots&x_{n-1}^{n}&x_n^{n}\end{pmatrix}$$

```python
## Answer
def VdM(x):
    n = x.size
    M = np.ones([n,n])
    for k in range(1,n):
        M[k,:] = M[k-1,:]*x
    return M
```

```python
x = np.array([1, 2, 4, 5])
print('Van der Monde matrix :\n', VdM(x))
```

<div markdown=1 class="DoIt"> 4°) In the nex cell, write a function with syntax `(interpVals,coeffs) = LagrangePol_vdm(x,y,X)` that uses the Van der Monde method to return
- the values `interpVals` of the Lagrange polynomial associated to $(x_k,y_k)_k$ evaluated at the points $X$ as its first output.
- the coefficients `coeffs` of the Lagrange polynomial as its second output.

Reminder: Use `np.transpose()`, to transpose a matrix `np.linalg.inv` for matrix inversion and `@` for matrix-vector product. 

```python
## Answer
def LagrangeInterp_vdm(x,y,X):
    M = VdM(x); M = M.transpose();
    coeffs = np.linalg.inv(M)@y
    interpVals = eval_Horner(coeffs,X)
    return (interpVals,coeffs)
```

<div markdown=1 class="DoIt"> 6°) Explain why the values of the Lagrange polynomial do not match *exactly* the target function at $(x_k)$.

<div markdown = 1, class = "Rmk">If $f$ is the function to be interpolated, $(x_k)_{0 \leq k \leq n}$ are the interpolation points used and $P_n$ is the corresponding Lagrange polynomial, we can quantify the quality of the approximation using the following error definition. 

$$e(f,(x_k)_k) = \sup_{[a,b]}\lvert f(x) - P_n(x) \rvert$$

We call this the **uniform error**. 

<div markdown = 1, class = "Def"> In the particular case where the points $(x_k)_{0 \leq k \leq n}$ are equispaced in $[a,b]$ with $a= x_0$ and $b = x_n$, we note

$$E_{\text{equi}}(n) = e(f,(x_k)_k) = \sup_{[a,b]}\lvert f(x) - P_n(x) \rvert$$

the uniform error of interpolation with equispaced points. For a given continuous function $f$, we say that the Lagrange polynomial of $f$ at the equispaced points converges **uniformly** towards $f$ on $[a,b]$ if 

$$E_{\text{equi}}(n) \to 0, \quad \text{as } n \to + \infty$$

Run the code in the next cell. It uses the functions you just coded to plot the evolution of $E_{\text{equi}}(n)$ as $n$ with $f$ taken as the sine wave.

```python
# Function that computes the Lagrange interpolation polynomial for f with equispaced n points on [a,b]
def Pn(a,b,f,n,X):
    x_interp = np.linspace(a,b,n)
    (res,temp) = LagrangeInterp_vdm(x_interp,f(x_interp),X)
    return res

# We restrict our attention to [-1,1]
a = -1; b = 1
Xtest = np.linspace(a,b,200)
ns = np.arange(1,45) 
Eequi = np.zeros(ns.size) # Pre-allocation
for n in ns:
    Eequi[n-1] = np.max(abs(Pn(a,b,sineWave,n,Xtest) - sineWave(Xtest)))

fig = plt.figure(figsize = (15,5))
plt.semilogy(ns,Eequi,marker='o')
plt.title('Uniform error of Equisaced Lagrange scheme',fontsize = 20)
plt.ylabel('$E_{equi}(n)$',fontsize = 20)
plt.xlabel('n',fontsize = 20)
plt.show()
```

<div  markdown=1 class="Rmk"> The previous plot has two parts. 
- In the first part, we see that the interpolation error decreases as $n$ increases, conformly with intuition - the more interpolation points, the better the approximation. 
- In the second part, one can see that the error goes which at first could seem surprising.

In the present case, this is due to rounding errors. It turns out that the Van der Monde method is very "ill-conditioned" meaning that a lot of rounding errors occur in the resolution of the linear system (some matrix are better conditioned than others). A lot of numerial algorithms have this kind of behaviour as the data size increases - that is the rounding errors become predominent. Such methods are sometimes termed "instable". 

<div  markdown=1 class="Rmk"> In the Van der Monde method, we need to perform the inversion of a $n\times n$ matrix. This requires $O(n^3)$ operations. The next method we will study only requires $O(n^2)$ operations to compute the coefficients. We will also see that it is a bit more stable. 

#### The divided differences method

<div  markdown=1 class="Def"> **Divided differences**

Let $(x_i,y_i)_{0 \leq i \leq N}$ a family of points. The divided differences are defined by the following recursive formula: 

\begin{align}
\delta^0y[x_i] &= y_i \\
\delta^1y[x_i,x_i+1] &= \frac{\delta^0y[x_{i+1}] - \delta^0y[x_i]}{x_{i+1} - x_i} \\
\delta^2y[x_i,x_{i+1},x_{i+2}] &= \frac{\delta^1y[x_{i+1},x_{i+2}] - \delta^1y[x_i,x_{i+1}]}{x_{i+2} - x_i} \\
\delta^ky[x_i,\cdots,x_{i+k}] &= \frac{\delta^{k-1}y[x_{i+1},\cdots,x_{i+k}] - \delta^{k-1}y[x_i,\cdots,x_{i+k-1}] }{x_{i+k} - x_i}
\end{align}

<div  markdown=1 class="Thm"> The Lagrange polynomial associated to $(x_k)_k$ and $(y_k)_k$ is given by 

$$P(X) = \delta^0y[x_0] + (x-x_0)\delta^1y[x_0,x_1] + (x-x_0)(x-x_1)\delta^2y[x_0,x_1,x_2] + \cdots + \\(x-x_0)(x-x_1)\cdots(x-x_{n-1})\delta^{n}y[x_0,\cdots,x_n].$$

- Pour la preuve, il y a peut-être besoin d'aborder quelques notions sur les polynômes (par exemple sur les coefficients dominants etc.)


> **Proof.** For $n=0$, the result is obvious. Let us assume that the result is true for $n-1$, let's show that it also holds for $n$. Let $P_n$ the Lagrange polynomial associated to $(x_k,y_k)_{0 \leq k \leq n}$ (remember $P_n$ is of degree at most $n$). Consider the Lagrange polynomials $P_{n-1}$ and $Q_{n-1}$ associated respectively to $(x_k,y_k)_{0 \leq k \leq n-1}$ and $(x_k,y_k)_{1 \leq k \leq n}$. Let $R_n$ and $S_n$ the two polynomials of degree at most $n$ such that 
>
>$$P_{n}(X) = P_{n-1}(X) + R_n(X) = Q_{n-1}(X) + S_n(X).$$
>
> By definition of the Lagrange polynomials, we see that $R_n(X) = P_{n+1}(X) - P_n(X)$ vanishes at $X = x_0, \cdots, x_{n-1}$, thus, since $R_n$ is of degree at most $n$, there exists a constant $a_R$ such that 
>
>$$R_n(X) = a_R(X-x_0)\cdots (X-x_{n-1}).$$
>
> Similarly, there exists a constant $a_S$ such that 
>
>$$S(X) = a_S(X-x_1)\cdots (X-x_{n}).$$
> 
> Now, observe that $S_n - R_n = Q_{n-1} - P_{n-1}$. The remainder of the proof just consists in identifying the coefficients of oerder $n$ and $n-1$ of these two polynomials. The right-hand side is of degree at most $n-1$ which implies that $S_n$ and $R_n$ must have the same coefficient of order $n$, that is 
>
> $$a_R = a_S,$$
> 
> which we now call $a$. We can thus write 
>
>$$S_n(X) - R_n(X) = a(X-x_1)\cdots(X-x_{n-1}) \left[(X - x_0) - (X-x_n)\right].$$
> 
> Matching the coefficient of order $n-1$ of $S_n - R_n$ with that of $P_{n-1} - Q_{n-1}$ now yields, according to the reccurence hypothesis:
>
>$$a(x_n - x_0) = \delta^{n-1}y[x_1,\cdots,x_n] - \delta^{n-1}y[x_0,\cdots,x_{n-1}].$$
> 
> Therefore $a = \delta^{n}y[x_0,\cdots,x_n]$ and using $P_n = P_{n-1} + R$, we get 
> 
> $$ P_n(X) =  \delta^0y[x_0] + (x-x_0)\delta^1y[x_0,x_1] + (x-x_0)(x-x_1)\delta^2y[x_0,x_1,x_2] + \cdots + \\(x-x_0)(x-x_1)\cdots(x-x_{n-1})\delta^{n}y[x_0,\cdots,x_n]$$
>
> as intended

Using this formula, we can again derive a modified Hörner scheme to evaluate the Lagrange polynomial, by rewriting: 

$$\begin{eqnarray} 
P(X) &=& \delta^0y[x_0] \\
&& + (X-x_0) \left( \delta^1y[x_0,x_1]\right.\\
&& + (X-x_1) \left(\delta^2y[x_0,x_1,x_2]\right.\\
&& + \cdots\\
&& + (X-x_{n-1})\left.\left. \delta^ny[x_0,\cdots,x_n]\right)\right)\end{eqnarray}$$

that allows to compute the value of $P$ at $X$ in $O(n)$ operations. That is, with this method, the cost for obtaining the coefficients is $0(n^2)$ (instead of $O(n^3)$ for the Van der Monde method), and $0(n)$ for evaluating the polynomial (as in the Van der Monde method). 

<div markdown=1 class="DoIt"> 7°) Write in the next cell a function with syntax `R = eval_Horner2(a,x,X)` implementing the modified Hörner method, that is, computing the polynomial 

$$\begin{eqnarray} 
P(X) &=& a_0 \\
&& + (X-x_0) \left( a_1\right.\\
&& + (X-x_1) \left( a_2\right.\\
&& + \cdots\\
&& + (X-x_{n-1})\left.\left. a_n \right)\right)\end{eqnarray}$$

(Be careful with the indices, $a$ and $x$ don't have the same length !)

```python
## Answer
def eval_Horner2(a,x,X):
    N = a.size
    R = a[-1]
    for k in range(1,N):
        R = R*(X - x[- k])
        R = R + a[-1 - k]
    return R
```

To compute the polynomial, we introduce an intermediray vector: 
$$\delta^k = (\delta^0y[x_0],\delta^1y[x_0,x_1],\delta^2y[x_0,x_1,x_2],\cdots,\delta^ky[x_0,\cdots,x_k],\delta^ky[x_1,\cdots,x_{k+1}],\dots,\delta^ky[x_{n-k},\cdots,x_n])$$

Note that the coefficients we are interested in are just given by $\delta^n$. Moreover, with this definition of $\delta^k$, we have in fact the following recursive property $\delta^0 = y$, $\forall i \leq k$, $\delta^{k+1}[i] = \delta^k[i]$, and for $i>k$, 

$$\delta^{k+1}[i] = \delta^{k+1}y[x_{i-k-1},\cdots,x_i] = \frac{\delta^{k}y[x_{i-k},\cdots,x_i] - \delta^k y[x_{i-k-1},\cdots,x_{i-1}]}{x[i] - x[i - k - 1]} = \frac{\delta^k[i] - \delta^k[i-1]}{x[i] - x[i-k-1]}$$

(Thus, at iteration k, we only have to modify the entries `delta[k:]`...)

<div markdown=1 class="DoIt"> 8°) Write a function with syntax `delta = divided_diff(x,y)` that returns the vector $\delta = (\delta^0y[x_0],\delta^1y[x_0,x_1],\cdots,\delta^n[x_0,x_1,\cdots,x_n])$ using the previous recursive formula. 

```python
# Answer 
# !! attention c'est pas hyer intuitif !!
def divided_diff(x,y):
    n = x.size
    deltak = y.copy() # Use copy to prevent the values in y from being modified 
    # Je pense qu'il faut laisser la ligne précédente - et il faut parler de ce concept de pointeurs en python.
    # Moi je me suis fait avoir. 
    for k in range(1,n):
        deltak[k:] = (deltak[k:] - deltak[k-1:-1])/(x[k:] - x[:-k])
    return deltak
```

- `r = eval_Horner2(delta,x[:-1],X)`

<div markdown=1 class="DoIt"> 9°) In the nex cell, write a function with syntax `(R,delta) = LagrangeInterp_dd(x,y,X)` that uses the divided differences to return

```python
## Answer
def LagrangeInterp_dd(x,y,X):
    delta = divided_diff(x,y)
    R = eval_Horner2(delta,x[:-1],X)
    return (R,delta)
```

#### The Chebyshev polynomials

To enhance the quality of the approximation, one way is to look for better nodes $(x_k)_k$. Recalling the estimation: 
$$f(y) - p(y) = (y-x_0)(y-x_1)\cdots(y-x_n) \frac{f^{(n+1)}(\xi)}{(n+1)!}$$
we could choose the points $(x_0, x_1, \cdots, x_n)$ so as to minimize the quantity

$$\sup_{y \in [a,b]}\lvert(y-x_0)(y-x_1)\cdots(y-x_n)\rvert.$$

It turns out that the choice of $n$ reals $(x_k)_k$  $n$ that achieves this minimum for [a,b] = [-1,1] is wel-known, and is related to the Chebyshev polynomials that we will now study. In fact, this problem was exactly what lead Chebyshev to introduce his polynomials...

<div  markdown=1 class="Def"> The Chebyshev polynomial of order $n$ is noted $T_n$ and is defined by recurrence by $T_0(X) = 1$, $T_1(X) = X$ and for all $n$
$$T_{n+1}(X) = 2XT_n(X) - T_{n-1}(X).$$

<div  markdown=1 class="DoIt"> 12°) Show by recurrence the following property: $\forall \theta \in \mathbb{R},\forall n \in \mathbb{N}$ 

$$T_n(\cos(\theta)) = \cos(n\theta).$$

Hint : apply trigonometric identities to $\cos((n+1)\theta) + \cos((n-1)\theta)$

Also prove by reccurence that the dominant coefficient of $T_n$ is $2^{n-1}$. 

Trigonometric identity:
$T_0 = 1 = \cos(0\theta)$. Let us assume the result is true for $k\leq n$, and show that it then holds for $k=n+1$. We write 
$$\cos((n+1)\theta)) + \cos((n-1)\theta)) = 2\cos(n\theta)\cos(\theta)$$
We conclude, using the recurrence hypothesis 
$$\cos((n+1)\theta) = 2\cos(\theta) T_n(\cos(\theta)) - T_{n-1}(\theta)$$
and the recursive definition of $T_n$: 
$$\cos((n+1)\theta) = T_{n+1}(\cos(\theta))$$
Dominant coefficient: 
Let us show by recurrence that $T_n$ is of degree $n$ with dominant coefficient $2^{n-1}$. For $n=0$, this is true. If it is true for all $k \leq n$, let's show it hold for $k = n+1$. We use 
$$T_{n+1}(X) = 2XT_{n}(X) - T_{n-1}(X).$$
In this equation, the right hand side is the sum of a polynomial of degree $n+1$ and a polynomial of degree $n-1$. Thus, it is of degree $n-1$. We conclude by identifying the dominant coefficients, using the recurrence hypothesis.  

<div  markdown=1 class="DoIt"> 13°) For all $0 \leq k \leq n-1$, let $\hat{x}_k = \cos\left(\frac{(2k + 1)\pi}{2n}\right)$. Show that $(\hat{x}_k)_k$ are $n$ distinct points in $]-1,1[$ with, for all $k \in [0,n-1]$

$$T_n(\hat{x}_k) = 0.$$

Deduce that 

$$T_n(X) = 2^{n-1}(X-\hat{x}_0)\cdots(X-\hat{x}_{n-1}).$$

Finally, show that $\lvert T_n \rvert$ reaches its maximal value (equal to $1$)  on [-1,1] exactly $n+1$ times, at the points $\cos\left(\frac{k\pi}{n}\right), k\in [0,n]$.

En fait plutôt (ce qui suffit) check that 
- $T_n\left[\cos\left(\frac{2k}{n}\pi\right)\right] = 1, \quad \forall 0\leq 2k \leq n$
- $T_{n}\left[\cos\left(\frac{2k+1}{n}\pi\right)\right] = -1, \quad \forall 0\leq 2k+1 \leq n$


Since $0 \leq k \leq n-1$, we have $0 < \frac{2k+1}{2n}\pi < \pi$. Let's call $\omega_k = \frac{(2k+1)\pi}{2n}$. Then we have $\hat{x}_k = \cos(\omega_k)$ and 

$$0 < \omega_0 < \omega_1 < \cdots < \omega_{n-1} < \pi.$$

Since the $\cos$ function is stricly decreasing on $]0,\pi[$, we deduce 

$$-1 < \hat{x}_{n-1} < \hat{x}_{n-2} < \cdots < \hat{x}_{0} < 1.$$

Therefore, $\hat{x}_k$ are $n$ distinct points in $]-1,1[$. To show that $T_n(\hat{x}_k) = 0$ for all $k$, we use the trigonometric property $T_n(\cos\theta) = \cos(n\theta)$:

$$T_n(\hat{x}_k) = \cos\left(n \frac{2k+1}{2n} \pi\right) = \cos\left(\frac{\pi}{2} + k\pi\right).$$

which vanishes for all $k$. 

Since $T_n$ is of degree $n$, we have now identified all of its roots (the $(\hat{x}_k)_k$) so there exists a constant $C$ such that 

$$T_n(X) = C \left(X - \hat{x}_0\right) \left(X - \hat{x}_1\right)\cdots \left(X - \hat{x}_{n-1}\right).$$

$C$ is the dominant coefficient of $T_n$ so according to the previous, $C = 2^{n-1}$. 

Maxima: (pas nécessaire en fait) We know that for all $X \in [-1,1]$, there is a unique $\theta \in [0,\pi]$ with $\cos(\theta) = X$. We will note this $\theta(X)$ for convenience. Thus, we have 

$$T_n(X) = T_n(\cos(\theta(X))) = \cos(n\theta(X))$$

which is bounded between $-1$ and $1$. Moreover, we have 

\begin{align}
T_n(X) = 1 &\iff T_n(\cos(\theta(X)) = 1 \\
&\iff \cos(n \theta(X)) = 1 \\
&\iff n\theta(X) = 0 [2\pi] \\
&\iff \exists k \text{ s.t. } 0 \leq 2k \leq n-1 : \theta(X) = \frac{2k\pi}{n}
\end{align}
Thus values $X$ at which $T_n(X) = 1$ are $\lbrace{X = \cos\left(\frac{2k}{n}\pi\right) | 0 \leq 2k \leq n\rbrace}$. Similarly, the values $X$ at which $T_n(X) = -1$ are $\lbrace{X = \cos\left(\frac{2k+1}{n}\pi\right) | 0 \leq 2k+1 \leq n\rbrace}$. This makes a total of $n+1$ real numbers $x$ at which $|T_n(x)| = 1$.

(Il suffit de se contenter de vérifier que $T_n\left(\cos\left(\frac{p\pi}{n}\right)\right) = \cos\left(n\frac{p \pi}{n}\right) = \cos(p\pi) = (-1)^p$.

<div  markdown=1 class="Thm"> Let $n \in \mathbb{N}$ and let $P_n$ be any polynomial of degree $n$ with a dominant coefficient equal to $1$. There holds

$$ \inf_{y \in [-1,1]}\lvert (y-\hat{x}_0)\cdots(y-\hat{x}_{n-1}) \rvert \leq \inf_{[-1,1]}\lvert P_n(y) \rvert.$$

In other words, among polynomials with unit dominant coefficient, $(X-\hat{x}_0)\cdots(X-\hat{x}_{n-1}) = \frac{T_n(X)}{2^{n-1}}$ is one of the "optimal" choices with respect to the infimum criterion.  

> **Proof.** Assume by contradiction that there exists a polynomial $P_n$ with degree $n$ such that 
>
> $$\inf_{y \in [-1,1]} \lvert P_n(y)\rvert < \inf_{y \in [-1,1]} \lvert (X-x_0)\cdots(X-x_{n-1})\rvert $$
> 
> that is, for all $y$, $\lvert P_n(y) \rvert < \frac{1}{2^{n-1}}$. Let $D_{n-1} = P_n -  (X-x_0)\cdots(X-x_{n-1})$ which is a polynomial of degree at most $n-1$ (since $P_n$ and $ (X-x_0)\cdots(X-x_{n-1})$ have the same leading coefficient, $1$). Notice that
> - $D_{n-1}\left[\cos\left(\frac{2k\pi}{n}\right)\right] < 0, \quad 0\leq 2k \leq n$
> - $D_{n-1}\left[\cos\left(\frac{2k+1\pi}{n}\right)\right] > 0, \quad 0\leq 2k+1 \leq n$
>
> By the intermediate value theorem, this implies that $D_n$ must vanish $n$ times. This is not possible since it is a (non-zero) polynomial of degree at most $n-1$. 

<div markdown = 1, class = "DoIt"> 13°) Write in the next cell a function with syntax `xhat(n)` that returns the array with values $(\hat{x}_0,\cdots, \hat{x}_n)$ sorted in ascending order. (use `np.sort`)

```python
## Answer 
def xhat(n):
    if n == 0:
        return np.array([])
    else:
        x = np.sort(np.cos(2*np.arange(0,n+1)/(2*n)*np.pi))
        return x
```

<div markdown = 1, class = "Rmk"> In theory, although the Chebyshev repartition of roots is better than the equispaced repartition, one can show that there are still some functions for which:

$$\lim_{n \to +\infty}E_{\text{cheb}}(n) = + \infty$$

This means that the Lagrange interpolation method does not aways succeed, even with the Chebyshev nodes.  

In fact, there is no repartition of points that would work for every function. That is, if for each $n$ we choose a set of $n$ points in $[a,b]$, there will always a function $f$ such that, letting $(P_n)_{n}$ the sequence of Lagrange interpolation polynomial of $f$ at these points, 

$$ \lim_{n \to +\infty} \sup_{x \in [a,b]} \lvert f(x) - P_n(x)\rvert = + \infty$$

This, with the numerical instability of the method, limits the range of applications of Lagrange interpolation. In many cases, on wish to have the guarantee that as the number of interpolation points increase, the quality of the approximation tends to zero. This is a feature of Piecewise interpolation that we study next.

## Piecewise polynomial interpolation

### Piecewise constant interpolation $\mathbb{P}^0$

<div markdown = 1 class = "Def"> Let $f$ a function defined on $[a,b]$ and a set of $n+1$ points $(x_k)_{0\leq k\leq n}$, with $x_0 = a$ and $x_n = b$. The piecewise constant interpolate of $f$ on $[a,b]$ with respect to the points $(x_k)_k$ is the function $\pi f$ defined on $[a,b]$ such that $\pi f(x) = f(x_k)$ for all $k < n$ and $x \in [x_k, x_{k+1}[$. 

<div markdown = 1, class = "Thm"> If $f$ is of class $C^1$, and if we note

$$h = \max_{k \in [0,n+1]}{\lvert x_{k+1} - x_k \rvert}$$

One has 

$$ \sup_{x \in [a,b]} \lvert \pi f(x) - f(x) \rvert \leq h \sup_{x\in [a,b]} |f'|$$

Preuve = accroissements finis.

```python
##Answer (particulièrement dur à coder, ils n'y arriveront pas). 

def nearestLeftNeighbour(x_interp,X): #replaces the elements Xi with the greatest xi such that xi <= Xi
    x_interp = x_interp[:-1]
    Y = X.copy()
    Y = np.sort(Y)
    y = x_interp.copy()
    y = np.sort(y)
    assert(y[0] <= Y[0]) # data in X must be within interpolation points
    j1 = 0
    j2 = 0
    n = y.size
    m = Y.size
    idx = np.zeros(m,dtype='int') # Grosse subtilité... il y en a besoin pour P1
    i=0
    while i < n-1 and j2 < m:
        if Y[j2] <= y[i+1]:
            j2 = j2+1
        else:
            Y[j1:j2] = y[i]
            idx[j1:j2] = i
            j1 = j2
            i = i+1
        
    Y[j1:] = y[i]
    idx[j1:] = i
    return (Y,idx)

def P0interpolate(f,x_interp,X):
    (r,tmp) = nearestLeftNeighbour(x_interp,X)
    return f(r)

f = sineWave
fig = plt.figure(figsize = (15,3))
x_interp = np.linspace(-1,1,50)
x_plot = np.linspace(-1,1,1000)
plt.plot(x_plot,f(x_plot))
plt.plot(x_plot,P0interpolate(f,x_interp,x_plot))
plt.legend(['function','P0 interpolate'])
plt.show()

#ajouter titres + axes
```

<div markdown = 1 class = "Def"> Let $f$ a function defined on $[a,b]$ and a set of $n+1$ points $(x_k)_{0\leq k\leq n}$, with $x_0 = a$ and $x_n = b$. The piecewise linear interpolate of $f$ on $[a,b]$ with respect to the points $(x_k)_k$ is the function $\pi f$ defined on $[a,b]$ such that $\pi f(x) = f(x_k) + \frac{x - x_k}{x_{k+1} - x_{k}}(f(x_{k+1}) - f(x_k))$ for $x \in [x_k, x_{k+1}[$.

```python
def nearestRightNeighbour(x_interp,X): #replaces the elements Xi with the smallest xi such that Xi < xi
    (tmp,idx) = nearestLeftNeighbour(x_interp,X)
    y = x_interp.copy()
    return y[idx + 1]

def P1interpolate(f,x_interp,X):
    (Xleft,idx) = nearestLeftNeighbour(x_interp,X)
    Xright= nearestRightNeighbour(x_interp,X)
    return f(Xleft) + (X - Xleft)/(Xright - Xleft)*(f(Xright) - f(Xleft))

f = sineWave
fig = plt.figure(figsize = (15,3))
x_interp = np.linspace(-1,1,50)
x_plot = np.linspace(-1,1,1000)
plt.plot(x_plot,f(x_plot))
plt.plot(x_plot,P1interpolate(f,x_interp,x_plot))
plt.legend(['function','P1 interpolate'])
plt.show()

# ajouter titres + axes
```

Let us check the order of the error as a function of $h = \frac{1}{n}$. (Here we only take equispaced points). 

```python
def E_piecewise(f,a,b,n,orderInterp):
    def Spline(f,n,X,orderInterp):
        x_interp = np.linspace(a,b,n)
        if orderInterp==0:
            res = P0interpolate(f,x_interp,X)
        elif orderInterp==1:
            res = P1interpolate(f,x_interp,X)
        return res

    Xtest = np.linspace(-1,1,np.max(n)*2 + 50)
    E_p = np.zeros(n.size) # Pre-allocation
    i = 0
    for ni in n:
        E_p[i] = np.max(abs(Spline(f,ni,Xtest,orderInterp) - f(Xtest)))
        i = i+1
    return E_p

nvec = np.arange(10,400)
h = 1/nvec
fig = plt.figure(figsize = (5,5))
plt.loglog(h,E_piecewise(sineWave,-1,1,nvec,0),marker = 'o',label = 'P0')
plt.loglog(h,E_piecewise(sineWave,-1,1,nvec,1),marker = 'o',label = 'P1')
plt.xlabel('h',fontsize = 20)
plt.ylabel('Uniform error',fontsize = 20)
plt.title('Piecewise polynomial interpolation, order of convergence',fontsize = 20)
plt.show()
```
