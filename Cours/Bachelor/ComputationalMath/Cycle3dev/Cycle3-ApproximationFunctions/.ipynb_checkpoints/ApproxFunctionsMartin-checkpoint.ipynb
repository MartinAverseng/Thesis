## Introduction

Suppose one wishes to compute the values $f(x)$ for a large number of arguments $x$, but that either:
- The function $f$ is not known explicitly (e.g. $f(t)$ is the average tempeature on earth at a time $t$...)
- It takes a long time to evaluate $f$ (for example when it is the output of a complicated code)

However, assume we already know the values $y_k = f(x_k)$ for $x_0, \cdots, x_N$. The arguments $(x_k)$ are called the **data points**. In practice, those values may come from experimental measures or previous runs of the code. 

In the interpolation method, we will approximate $f(x)$ for any $x$ by

$$f(x) \approx P(x)$$

where $P$ is a function that **exactly agrees** with $f$ at every data point $(x_k)$, that is, satisfying $P(x_k) = f(x_k)$ for all $0 \leq k \leq N$. 

- If $P$ is chosen as the polynomial of minimal degree satisfying this condition, this is the Lagrange interpolation method. 
- If $P$ is chosen such its restriction to each interval $]x_k,x_{k+1}[$ is a polynomial, this is piecewise polynomial interpolation. 

In the least-square method, we will approximate $f(x)$ for any $x$ by a model of the form 

$$f(x) \approx \beta_0 + \beta_1 x + \cdots + \beta_N x^N $$

For example affine regression is a least-square method whith two parameters $\beta_0$ and $\beta_1$, and with

$$f(x) \approx \beta_0 + \beta_1 x.$$

Contrary to the interpolation approach, one does not enforce exact agreement between the model polynomial and  $f$ at $x_k$ (because in general we use a polynomial of small degree). However, the values of the parameters $\beta_i$ are chosen so as to maximize this agreement (or, equivalently, to minimize the error). That is, we use the known data $(x_k)$ and $(y_k)$ to "tune" the parameters $(\beta_i)$ to be used later for new  unknown values $x$.

## Lagrange interpolation

### Introduction

In this first section, we choose to approximate the function $f$ by a polynomial $P$. We will answer questions such as: how to choose the degree of this polynomial ? Is it unique ? How can we efficiently compute its coefficients from the knowledge of $(y_k)$ and $(x_k)$ ? Under which conditions is the error $\lvert P(x) - f(x) \rvert$ sufficiently small ? If we are free to choose the points $(x_k)$, is there a particularly good choice ?

### Interpolation polynomial of minimal degree

#### Existence and unicity of the Lagrange polynomial

<div  markdown=1 class="Thm">  Let $x_0, x_1, \cdots, x_n \in [a,b]$ a set of **interpolation points** and $y_0,y_1,\cdots,y_n$ a set of arbitrary **real numbers**. There exists a unique real polynomial $p_n$ of degree less than $n$ such that, $p_n(x_i) = y_i$ for all $0 \leq i \leq n.$ This polynomial is explicitly given by 

$$p_n(X) = \sum_{i=0}^n y_i L_i(X)$$

where $(L_i)_i$ are the elementary *Lagrange interpolation* polynomial given by 

$$ L_i(X) = \prod_{j \neq i}\frac{X - x_j}{x_i - x_j}, \quad 0 \leq i \leq n$$

#### Error bound

<div  markdown=1 class="Thm"> De Rolle

<div  markdown=1 class="Thm"> De Rolle généralisé

<div  markdown=1 class="Thm">  Let $f : [a,b] \to \mathbb{R}$ be $n+1$ times differentiable and let $P_n$ the Lagrange polynomial associated with the points $(x_i, f(x_i))_{0\leq i \leq n}$. Then, for every $y$ in $[a,b]$, there exists $\xi \in ]a,b[$ such that 

$$f(y) - p(y) = (y-x_0)(y-x_1)\cdots(y-x_n) \frac{f^{(n+1)}(\xi)}{(n+1)!}$$

<div  markdown=1 class="Thm">  Using the same notations as in the previous theorem, we have, for all $y \in [a,b]$, 

$$\lvert f(y) - p(y) \rvert \leq \frac{(b-a)^n}{n!} \sup_{[a,b]} \lvert f^{(n+1)} \rvert $$

### Computation of the Lagrange polynomai

#### Preliminaries : Hörner algorithm

In this section, we will need to evaluate polynomials of the form 

$$P(X) = a_0 + a_1 X + \cdots + a_N X^N$$

with given coefficients $(a_0,\cdots,a_N)$. For this, we rewrite 

$$P(X) = a_0 + X(a_1 + X ( a_2 + X ( \cdots ))))$$

Leading to the Hörner algorithm for evaluating the polynomial

<div  markdown=1 class="Algo">
**Hörner algorithm.** Evaluates  a polynomial with coefficients $(a_k)_k$ at $X$.


\begin{align}
INPUT:&\quad (a_k)_{0\leq k \leq N-1}, X\\
INIT:
&\quad R = a_{N-1}\\
FOR: &\quad k = 1 .. N-1 \\
&\quad\quad R = R X + a_{N-1-k}\\
RETURN:&\quad R\\
\end{align}

<div markdown=1 class="DoIt"> 1°) Write in the next cell a function with syntax `R = eval_Horner(a,X)` implementing the previous algorithm. 

```python
## Answer
def eval_Horner(a,X):
    N = a.size
    R = a[-1]
    for k in range(1,N):
        R = R*X + a[-1 -k]
    return R
```

Test your function by evaluating the following cell. 

```python
a = np.array([1,2,3]) # P(X) = 1 + 2X + 3X^2
x = np.array([0, 1, 2, 3, 6]) 
solution = np.array([1,6,17,34,121])
R = eval_Horner(a,x)
print('your code output :', R)
print('The solution :', solution)
if np.allclose(solution,R):
    print(':-)')
else:
    print(':-()')
```

<div markdown=1 class="DoIt"> 2°) Define in the next cell a function with syntax `M = VdM(x)`, taking in argument an array of real values and computing the vandermonde matrix 

$$\begin{pmatrix} 1&1&\cdots & 1 &1\\x_0&x_1&\cdots &x_{n-1}&x_{n}\\\vdots&\vdots&\ddots&\vdots&\vdots\\x_0^{n-1}&x_1^{n-1}&\cdots&x_{n-1}^{n-1}& x_n^{n-1}\\x_0^{n}&x_1^{n}&\cdots&x_{n-1}^{n}&x_n^{n}\end{pmatrix}$$

```python
## Answer
def VdM(x):
    n = x.size
    M = np.ones([n,n])
    for k in range(1,n):
        M[k,:] = M[k-1,:]*x
    return M
```

```python
x = np.array([1, 2, 4, 5])
print('Van der Monde matrix :\n', VdM(x))
```

<div markdown=1 class="DoIt"> 4°) In the nex cell, write a function with syntax `(R,a) = LagrangePol_vdm(x,y,X)` that uses the Van der Monde method to return
- the values of the Lagrange polynomial associated to $(x_k,y_k)_k$ evaluated at the points $X$ as its first output.
- the coefficients of the Lagrange polynomial as its second output.

Reminder: Use `np.transpose()`, to transpose a matrix `np.linalg.inv` for matrix inversion and `@` for matrix-vector product. 

```python
## Answer
def LagrangeInterp_vdm(x,y,X):
    M = VdM(x); M = M.transpose();
    a = np.linalg.inv(M)@y
    R = eval_Horner(a,X)
    return (R,a)
```

<div  markdown=1 class="Rmk"> There are several drawbacks to the previous methods. First, the inversion of a $n\times n$ matrix is performed in $O(n^3)$ iterations. As we shall see, there are more efficient ways for obtaining the coefficients. Moreover, the Van der Monde matrix is generally "ill-conditionned", meaning that a lot of rounding errors occur in the resolution of the linear system. We will now introduce a better method, called the divided difference or Newton's method.

#### The divided differences method

<div  markdown=1 class="Def"> **Divided differences**

Let $(x_i,y_i)_{0 \leq i \leq N}$ a family of points. The divided differences are defined by the following recursive formula: 

\begin{align}
\delta^0y[x_i] &= y_i \\
\delta^1y[x_i,x_i+1] &= \frac{\delta^0y[x_{i+1}] - \delta^0y[x_i]}{x_{i+1} - x_i} \\
\delta^2y[x_i,x_{i+1},x_{i+2}] &= \frac{\delta^1y[x_{i+1},x_{i+2}] - \delta^1y[x_i,x_{i+1}]}{x_{i+2} - x_i} \\
\delta^ky[x_i,\cdots,x_{i+k}] &= \frac{\delta^{k-1}y[x_{i+1},\cdots,x_{i+k}] - \delta^y{k-1}[x_i,\cdots,x_{i+k-1}] }{x_{i+k} - x_i}
\end{align}

<div  markdown=1 class="Thm"> The Lagrange polynomial associated to $(x_k)_k$ and $(y_k)_k$ is given by 

$$P(X) = \delta^0y[x_0] + (x-x_0)\delta^1y[x_0,x_1] + (x-x_0)(x-x_1)\delta^2y[x_0,x_1,x_2] + \cdots + \\(x-x_0)(x-x_1)\cdots(x-x_{n-1})\delta^{n}y[x_0,\cdots,x_n].$$

> **Proof.** For $n=0$, the result is obvious. Let us assume that the result is true for $n-1$, let's show that it also holds for $n$. Let $P_n$ the Lagrange polynomial associated to $(x_k,y_k)_{0 \leq k \leq n}$ (remember $P_n$ is of degree at most $n$). Consider the Lagrange polynomials $P_{n-1}$ and $Q_{n-1}$ associated respectively to $(x_k,y_k)_{0 \leq k \leq n-1}$ and $(x_k,y_k)_{1 \leq k \leq n}$. Let $R_n$ and $S_n$ the two polynomials of degree at most $n$ such that 
>
>$$P_{n}(X) = P_{n-1}(X) + R_n(X) = Q_{n-1}(X) + S_n(X).$$
>
> By definition of the Lagrange polynomials, we see that $R_n(X) = P_{n+1}(X) - P_n(X)$ vanishes at $X = x_0, \cdots, x_{n-1}$, thus, since $R_n$ is of degree at most $n$, there exists a constant $a_R$ such that 
>
>$$R_n(X) = a_R(X-x_0)\cdots (X-x_{n-1}).$$
>
> Similarly, there exists a constant $a_S$ such that 
>
>$$S(X) = a_S(X-x_1)\cdots (X-x_{n}).$$
> 
> Now, observe that $S_n - R_n = Q_{n-1} - P_{n-1}$. The remainder of the proof just consists in identifying the coefficients of oerder $n$ and $n-1$ of these two polynomials. The right-hand side is of degree at most $n-1$ which implies that $S_n$ and $R_n$ must have the same coefficient of order $n$, that is 
>
> $$a_R = a_S,$$
> 
> which we now call $a$. We can thus write 
>
>$$S_n(X) - R_n(X) = a(X-x_1)\cdots(X-x_{n-1}) \left[(X - x_0) - (X-x_n)\right].$$
> 
> Matching the coefficient of order $n-1$ of $S_n - R_n$ with that of $P_{n-1} - Q_{n-1}$ now yields, according to the reccurence hypothesis:
>
>$$a(x_n - x_0) = \delta^{n-1}y[x_1,\cdots,x_n] - \delta^{n-1}y[x_0,\cdots,x_{n-1}].$$
> 
> Therefore $a = \delta^{n}y[x_0,\cdots,x_n]$ and using $P_n = P_{n-1} + R$, we get 
> 
> $$ P_n(X) =  \delta^0y[x_0] + (x-x_0)\delta^1y[x_0,x_1] + (x-x_0)(x-x_1)\delta^2y[x_0,x_1,x_2] + \cdots + \\(x-x_0)(x-x_1)\cdots(x-x_{n-1})\delta^{n}y[x_0,\cdots,x_n]$$
>
> as intended

Using this formula, we can again derive a modified Hörner scheme to evaluate the Lagrange polynomial, by rewriting: 

$$ P(X) = \delta^0y[x_0] + (X-x_0) \left( \delta^1y[x_0,x_1] + (X-x_1)\left(\delta^2y[x_0,x_1,x_2] + \cdots + (X-x_{n-1}) \delta^ny[x_0,\cdots,x_n]\right)\right)$$

that allows to compute the value of $P$ at $X$ in $O(n)$ operations. That is, with this method, the cost for obtaining the coefficients is $0(n^2)$ (instead of $O(n^3)$ for the Van der Monde method), and $0(n)$ for evaluating the polynomial (as in the Van der Monde method). 

<div markdown=1 class="DoIt"> 7°) Write in the next cell a function with syntax `R = eval_Horner2(a,x,X)` implementing the modified Hörner method, that is, computing the polynomial 

$$ P(X) = a_0 + (X-x_0) \left( a_1 + (X-x_1)\left(a_2 + \cdots + (X-x_{n-1}) a_n\right)\right)$$

(Be careful with the indices, $a$ and $x$ don't have the same length !)

```python
## Answer
def eval_Horner2(a,x,X):
    N = a.size
    R = a[-1]
    for k in range(1,N):
        R = R*(X - x[- k])
        R = R + a[-1 - k]
    return R
```

<div markdown=1 class="DoIt"> 8°) Write a function with syntax `delta = divided_diff(x,y)` that returns the vector $\delta = (\delta^0y[x_0],\cdots,\delta^n[x_0,\cdots,x_n])$. Write the function so that, after the $k-th$ iteration, the vector $\delta$ is given by

$$\delta = (\delta^0y[x_0],\delta^1y[x_0,x_1],\delta^2y[x_0,x_1,x_2],\cdots,\delta^ky[x_0,\cdots,x_k],\delta^ky[x_1,\cdots,x_{k+1}],\dots,\delta^ky[x_{N-k},\cdots,x_N])$$

(Note that at iteration k, we only modify the entries `delta[k:]`...)

```python
# Answer 
# !! attention c'est pas hyer intuitif !!
def divided_diff(x,y):
    n = x.size
    delta = y.copy() # Use copy to prevent the values in y from being modified 
    # Je pense qu'il faut laisser la ligne précédente - et il faut parler de ce concept de pointeurs en python.
    # Moi je me suis fait avoir. 
    for k in range(1,n+1):
        delta[k:] = (delta[k:] - delta[k-1:-1])/(x[k:] - x[:-k])
    return delta   
```

#### The Chebyshev polynomials

To temper the instability of the approximation, one way is to choose a better repartition of the nodes $(x_k)_k$. In particular, it looks like one should take more points near the edges of the interval. If we recall the estimation: 
$$f(y) - p(y) = (y-x_0)(y-x_1)\cdots(y-x_n) \frac{f^{(n+1)}(\xi)}{(n+1)!}$$
we could choose the points $(x_0, x_1, \cdots, x_n)$ so that the maximal value of the polynomial 
$$P(y) = (y-x_0)(y-x_1)\cdots(y-x_n)$$
involved in the previous estimation be minimal. 

It turns out that the polynomial of order $n$ that minimizes this norm is well-known and is called the "Chebyshev polynomial", and denoted by $T_n(X)$. 

<div  markdown=1 class="Def"> The Chebyshev polynomial of order $n$ is noted $T_n$ and is defined by recurrence by $T_0(X) = 1$, $T_1(X) = X$ and for all $n$
$$T_{n+1}(X) = 2XT_n(X) - T_{n-1}(X).$$

<div  markdown=1 class="Thm"> Let $n \in \mathbb{N}$ and let $P_n$ be any polynomial of degree $n$ with a dominant coefficient equal to $1$. There holds

$$ \inf_{y \in [-1,1]}\lvert (X-x_0)\cdots(X-x_{n-1}) \rvert \leq \inf_{[-1,1]}\lvert P_n(y) \rvert.$$

In other words, among polynomials with unit dominant coefficient, $(X-x_0)\cdots(X-x_{n-1}) = \frac{T_n(X)}{2^{n-1}}$ is one of the "optimal" choices with respect to the infimum criterion.  

> **Proof.** Assume that there exists a polynomial $P_n$ with degree $n$ such that 
>
> $$\inf_{y \in [-1,1]} \lvert P_n(y)\rvert < \inf_{y \in [-1,1]} \lvert (X-x_0)\cdots(X-x_{n-1})\rvert $$
> 
> that is, for all $y$, $\lvert P_n(y) \rvert < \frac{1}{2^{n-1}}$. Let $D_{n-1} = P_n -  (X-x_0)\cdots(X-x_{n-1})$ which is a polynomial of degree at most $n-1$ (since $P_n$ and $ (X-x_0)\cdots(X-x_{n-1})$ have the same leading coefficient, $1$). Notice that
> - $D_{n-1}\left[\cos\left(\frac{2k\pi}{n}\right)\right] < 0, \quad 0\leq 2k \leq n$
> - $D_{n-1}\left[\cos\left(\frac{2k+1\pi}{n}\right)\right] > 0, \quad 0\leq 2k+1 \leq n$
>
> This implies that $D_n$ must vanish $n$ times. This is not possible since it is a polynomial of degree at most $n-1$.
