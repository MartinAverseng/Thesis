```python
## loading python libraries

# necessary to display plots inline:
%matplotlib inline   

# load the libraries
import matplotlib.pyplot as plt # 2D plotting library
import numpy as np              # package for scientific computing  

from math import *              # package for mathematics (pi, arctan, sqrt, factorial ...)
```

```python
### function computing the zeros of Tn
def xhat(n):
    if n == 0:
        return np.array([])
    else:
        x = np.sort(np.cos(2*np.arange(0,n+1)/(2*n)*np.pi))
        return x
```

# How to compute $\pi$

One can use the following interpolation methods to compute integrals. However, the precision and the stability of the interpolation discussed earlier are essential to get correct results. This is illustrated here with the computation of $\pi$.

We know that $\forall \alpha\in\mathbb{R}_+^*$,$$\frac{\pi}{2}= \arctan(\alpha) + \arctan\left(\frac{1}{\alpha}\right)\qquad\textrm{and}\qquad \arctan(\alpha) = \int_{0}^{1}\frac{\alpha dx}{1+\alpha^2 x^2}dx$$

Thus, 

$$\frac{\pi}{2}=\int_{}^{1}\Big[\frac{\alpha}{1+\alpha^2 x^2}+\dfrac{\alpha}{\alpha^2 + x^2}\Big]dx$$

We thus set 

$$ d_{atan}: \left\{\begin{array}{l}
\mathbb{R} &\rightarrow &\mathbb{R} \\
x &\rightarrow &\dfrac{\alpha}{1+\alpha^2 x^2} + \dfrac{\alpha}{\alpha^2 + x^2} \end{array}\right.$$

and we get $$\frac{\pi}{2}=\int_{0}^{1} d_{atan}(x)dx$$
We want to approximate $d_{atan}$ by the previously seen methods and then to integrate. Three methods have been used and will be compared in the following :
- $\mathbb{P}^0$-interpolation
- $\mathbb{P}^1$-interpolation
- Lagrange Interpolation


```python
vstar=np.pi


def datan(x,alpha):
    return (alpha/(1+(alpha*x)**2) + 1/alpha*1/(1+(x/alpha)**2))
```

```python
# Get the integral of the P0-approximation corresponding to equidistant interpolation points
#    f : function to integrate
#    a : lower bound of the integral
#    b : upper bound of the integral
#    n : number of interpolation points
def intP0Equi(f,alpha,a,b,n):
    x=np.linspace(a,b,n)
    y=f(x,alpha)
    return(np.sum(y[:-1:])*(b-a)/(n-1))
    
```

```python
# Get the integral of the P1-approximation corresponding to equidistant interpolation points
#    f : function to integrate
#    a : lower bound of the integral
#    b : upper bound of the integral
#    n : number of interpolation points
def intP1Equi(f,alpha,a,b,n):
    x=np.linspace(a,b,n)
    y=f(x,alpha)
    squarePart=np.sum(y[:-1:])*(b-a)/(n-1)
    trianglePart=np.sum(y[1:]-y[:-1:])*(b-a)/(2*(n-1))
    return(squarePart+trianglePart)
    
```

```python
def VdM(x):
    n = x.size
    M = np.ones([n,n])
    for k in range(1,n):
        M[k,:] = M[k-1,:]*x
    return M
```

```python
def LagrangeInterp_vdm(x,y):
    M = VdM(x); M = M.transpose();
    coeffs = np.linalg.inv(M)@y
    return (coeffs)
```

```python
# Get the integral of the Lagrange-approximation corresponding to equidistant interpolation points
#    f : function to integrate
#    a : lower bound of the integral
#    b : upper bound of the integral
#    n : number of interpolation points

def intLagrangeEqui (f,apha,a,b,n):
    x=np.linspace(-1,1,n)
    y=f(x,alpha)
    coefs=LagrangeInterp_vdm(x,y)
    power=np.arange(coefs.size)
    aux=coefs*(b**(power+1)-a**(power+1))/(power+1)
    return(np.sum(aux))
```

```python
# Get the integral of the Lagrange-approximation corresponding to Chebychev interpolation points
#    f : function to integrate
#    a : lower bound of the integral
#    b : upper bound of the integral
#    n : number of interpolation points

def intLagrangeCheb (f,alpha,a,b,n):
    x=xhat(n)
    x=(x+1)/2
    y=f(x,alpha)
    coefs=LagrangeInterp_vdm(x,y)
    power=np.arange(coefs.size)
    aux=coefs*(b**(power+1)-a**(power+1))/(power+1)
    return(np.sum(aux))
```

```python
alpha=1;


a=0
b=1

# Compute the integral and the error for different values of n
nMax=50
ns=np.arange(nMax+1)


resIntP0Equi=np.zeros(ns.size-2)
resIntP1Equi=np.zeros(ns.size-2)
resIntLagrangeEqui=np.zeros(ns.size-2) 
resIntLagrangeCheb=np.zeros(ns.size-2)

for n in ns[:-2]:
    resIntP0Equi[n]=intP0Equi(datan,alpha,a,b,n+2)
    resIntP1Equi[n]=intP1Equi(datan,alpha,a,b,n+2) 
    resIntLagrangeEqui[n]=intLagrangeEqui(datan,alpha,a,b,n+2)
    resIntLagrangeCheb[n]=intLagrangeCheb(datan,alpha,a,b,n+2)
    
errP0Equi=abs(2*resIntP0Equi-vstar)
errP1Equi=abs(2*resIntP1Equi-vstar)
errLagrangeEqui=abs(2*resIntLagrangeEqui-vstar)
errLagrangeCheb=abs(2*resIntLagrangeCheb-vstar)

#print(errP0Equi)
#print(errP1Equi)
#print(errLagrangeEqui)

# distance entre les points d'interpolation
h=(b-a)/(ns[2:])

fig = plt.figure(figsize=(20, 13))
plt.loglog(h,errP0Equi, marker="x", label="P0") #log-log scale
plt.loglog(h,errP1Equi, marker="x", label="P1") #log-log scale
plt.loglog(h,errLagrangeEqui, marker="x", label="Lagrange, points equidistants") #log-log scale
plt.loglog(h,errLagrangeCheb, marker="x", label="Lagrange, points de Chebychev") #log-log scale
plt.title('Order of Convergence', fontsize = 24)
plt.xlabel('$h$',fontsize = 20)
plt.ylabel('$e_k$',fontsize = 20)
plt.legend(loc='best', fontsize = 18)

plt.show()
```
