```python
### function computing the zeros of Tn
def xhat(n):
    if n == 0:
        return np.array([])
    else:
        x = np.sort(np.cos(2*np.arange(0,n+1)/(2*n)*np.pi))
        return x
```

# How to compute $\pi$

One can use the following interpolation methods to compute integrals. However, the precision and the stability of the interpolation discussed earlier are essential to get correct results. This is illustrated here with the computation of $\pi$.

We know that $$\pi=4\int_{0}^{1}\frac{1}{1+x^2}dx $$

We set $$ d_{atan}:\left\{\begin{array}{l}
                           \mathbb{R} &\rightarrow &\mathbb{R} \\
                            x &\rightarrow &\frac{1}{1+x^2}
                            \end{array}\right. $$

We want to approximate $d_{atan}$ by the previously seen methods and then to integrate. Three methods have been used and will be compared in the following :
- $\mathbb{P}^0$-interpolation
- $\mathbb{P}^1$-interpolation
- Lagrange Interpolation

```python
def datan(x):
    return(1/(1+x**2))
```

Equidistant points are used here. The following functions are the computation of the integral for the different methods. To get the integral of Lagrange interpolation, we compute directly the coefficient of the polynom with the Van der Monde method.

```python
def VdM(x):
    n = x.size
    M = np.ones([n,n])
    for k in range(1,n):
        M[k,:] = M[k-1,:]*x
    return M
```

```python
def LagrangeInterp_vdm(x,y):
    M = VdM(x); M = M.transpose();
    coeffs = np.linalg.inv(M)@y
    return (coeffs)
```

```python
def intP0EQUI (nmax):
    ns=np.arange(2,nmax)
    inte=np.zeros(ns.size)
    for n in ns:
        x=np.linspace(0,1,n)
        y=datan(x)
        inte[n-2]=4*np.sum(y[:-1:])/(n-1)
    return(inte)

def intP1EQUI (nmax):
    ns=np.arange(2,nmax)
    inte=np.zeros(ns.size)
    for n in ns:
        x=np.linspace(0,1,n)
        y=datan(x)
        inte[n-2]=4*((np.sum(y[:-1:])+(np.sum(y[1:]-y[:-1:])/2))/(n-1))
    return(inte)

def intLagrangeEQUI (nmax):
    ns=np.arange(2,nmax)
    inte=np.zeros(ns.size)
    for n in ns:
        x=np.linspace(0,1,n)
        y=datan(x)
        coefs=LagrangeInterp_vdm(x,y)
        power=np.arange(coefs.size)
        aux=coefs/(power+1)
        inte[n-2]=4*np.sum(aux)
    return(inte)

def intLagrangeCHEB (nmax):
    ns=np.arange(1,nmax-1)
    inte=np.zeros(ns.size)
    for n in ns:
        x=xhat(n)
        x=x[x>0]
        y=datan(x)
        coefs=LagrangeInterp_vdm(x,y)
        power=np.arange(coefs.size)
        aux=coefs/(power+1)
        inte[n-1]=4*np.sum(aux)
    return(inte)
```

The errors are then plotted to see which method is the best.

```python
n=1000
# ERROR FOR THE P0 INTERPOLATION 
resP0EQUI=intP0EQUI(100)
#print(resP0EQUI)
errP0EQUI=abs(resP0EQUI-np.pi)
#print(errP0EQUI)

# ERROR FOR THE P1 INTERPOLATION 
resP1EQUI=intP1EQUI(100)
#print(resP1EQUI)
errP1EQUI=abs(resP1EQUI-np.pi)
#print(errP1EQUI)
    
# ERROR FOR THE LAGRANGE INTERPOLATION EQUI
resLAGRANGEEQUI=intLagrangeEQUI(100)
#print(resLAGRANGEEQUI)
errLAGRANGEEQUI=abs(resLAGRANGEEQUI-np.pi)
#print(errLAGRANGEEQUI) 
    
# ERROR FOR THE LAGRANGE INTERPOLATION CHEB
resLAGRANGECHEB=intLagrangeCHEB(100)
#print(resLAGRANGEEQUI)
errLAGRANGECHEB=abs(resLAGRANGECHEB-np.pi)
#print(errLAGRANGEEQUI) 

tab=np.arange(0,resP0EQUI.size)
#print(tab)


fig = plt.figure(figsize=(20, 13))
plt.plot(tab, errP0EQUI, marker="x", label="ErrorP0 vs iteration")
plt.plot(tab, errP1EQUI, marker="x", label="ErrorP1 vs iteration")
plt.plot(tab, errLAGRANGEEQUI, marker="x", label="Error Lagrange equi vs iteration")
plt.plot(tab, errLAGRANGECHEB, marker="x", label="Error Lagrange cheb vs iteration")
plt.yscale("log")
plt.title('Convergence', fontsize = 24)
plt.xlabel('$k$',fontsize = 20)
plt.ylabel('$e_k$',fontsize = 20)
plt.legend(loc='best', fontsize = 18)

#plt.subplot(122) #
#plt.loglog(errP0EQUI[:len(errP0EQUI)-1],errP0EQUI[1:], marker="x", label="Slope for P0") #log-log scale
#plt.loglog(errP1EQUI[:len(errP1EQUI)-1],errP1EQUI[1:], marker="x", label="Slopefor P1") #log-log scale
#plt.axis('equal')
#plt.title('Order of Convergence', fontsize = 24)
#plt.xlabel('$e_k$',fontsize = 20)
#plt.ylabel('$e_{k+1}$',fontsize = 20)
#plt.legend(loc='best', fontsize = 18)

plt.show()
```
