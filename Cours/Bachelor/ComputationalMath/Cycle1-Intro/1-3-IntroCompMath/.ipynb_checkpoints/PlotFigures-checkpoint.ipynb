# Approx Pi

```python
# necessary to display plots inline:
%matplotlib inline   

# load the libraries
import matplotlib.pyplot as plt # 2D plotting library
import numpy as np              # package for scientific computing  

from math import *
```

```python
## Function that computes the approximation of pi using archimedes method
## input : K = parameter (number of sides of the polygon = 6*2^K )
## output : P = vector of length K+1 
##              containing the list of approximations from k=0 to k=K
##              P(k) approximation using a k-sided polygon

def Archimede(K):
    # create vector P
    P = np.zeros(K+1)
    # P_0(x)=1/sqrt(3)
    t= 1/sqrt(3)
    P[0] = 6 * t
    # computation of P_k for k=1..N
    # using P_{k+1}(x) = P_{k}(x) + (-1)^{k+1}x^{2k+3}/(2k+3)
    for k in np.arange(K):
        t=t/(1+sqrt(t**2+1))
        P[k+1] = 6 * 2**(k+1) * t
    return P
```

```python
# approximation of pi
k=5
PI=Archimede(k)

# point (xC,yC) to plot the circle
N=100
thetaC=2*pi/N
tabthetaC = thetaC*np.arange(N+1)
tabxC=np.cos(tabthetaC)
tabyC=np.sin(tabthetaC)

k=0
#points (x,y) to plot the polygon
Nsides= 6*2**k
theta=2*pi/Nsides
tabtheta = theta*np.arange(Nsides+1)
tabx1=np.cos(tabtheta)
taby1=np.sin(tabtheta)

k2=1
#points (x,y) to plot the polygon
Nsides= 6*2**k2
theta=2*pi/Nsides
tabtheta = theta*np.arange(Nsides+1)
tabx2=np.cos(tabtheta)
taby2=np.sin(tabtheta)

k3=5
#points (x,y) to plot the polygon
Nsides= 6*2**k3
theta=2*pi/Nsides
tabtheta = theta*np.arange(Nsides+1)
tabx3=np.cos(tabtheta)
taby3=np.sin(tabtheta)

#figure
fig, axs = plt.subplots(1, 3,figsize=(30,10))

axs[0].plot(tabxC,tabyC)
axs[0].plot(tabx1,taby1,marker="o",label='k ='+str(k)+', Polygon with '+str(Nsides)+' sides')
#plt.legend(loc='upper left', fontsize=15)
#axs[0].set_aspect('equal', 'box')
axs[0].set_aspect('equal', 'box')
#axs[0].text(-0.6,0.25,'$\pi$ ='+str(pi),fontsize=18) #writes the exact value of pi
axs[0].text(-0.4,0,str(PI[k]),fontsize=30) #writes the computed approximation
axs[0].set_xticks([])
axs[0].set_yticks([])

axs[1].plot(tabxC,tabyC)
axs[1].plot(tabx2,taby2,marker="o",label='k ='+str(k)+', Polygon with '+str(Nsides)+' sides')
#plt.legend(loc='upper left', fontsize=15)
axs[1].set_aspect('equal', 'box')
#axs[1].text(-0.6,0.25,'$\pi$ ='+str(pi),fontsize=18) #writes the exact value of pi
axs[1].text(-0.5,0,str(PI[k2]),fontsize=30) #writes the computed approximation
axs[1].set_xticks([])
axs[1].set_yticks([])

axs[2].plot(tabxC,tabyC)
axs[2].plot(tabx3,taby3,marker="o",label='k ='+str(k)+', Polygon with '+str(Nsides)+' sides')
#plt.legend(loc='upper left', fontsize=15)
axs[2].set_aspect('equal', 'box')
#axs[2].text(-0.6,0.25,'$\pi$ ='+str(pi),fontsize=18) #writes the exact value of pi
axs[2].text(-0.5,0,str(PI[k3]),fontsize=30) #writes the computed approximation
axs[2].set_xticks([])
axs[2].set_yticks([])

plt.show()
```
