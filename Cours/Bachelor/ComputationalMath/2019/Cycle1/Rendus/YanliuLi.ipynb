```python
import numpy as np
a = np.random.rand(5,5) 
print('a =', a)
a_max = np.argmax(a, axis = 0)
a_min = np.argmin(a, axis = 0)
a[a_max], a[a_min] = a[a_min], a[a_max]
print('a =', a)
```

```python
a = np.random.rand(10,2) 
c = np.zeros((a.shape))
for i in range(np.shape(a)[0]):
    r = (a[i][0]**2 + a[i][1]**2)**(1/2)
    theta = np.arctan(a[i][1]/a[i][0])
    c[i] = [r, theta]
print('a =', a)
print('c =', c)
```

```python
%matplotlib inline   

# load the libraries
import matplotlib.pyplot as plt # 2D plotting library
import numpy as np    
```

```python
X = np.linspace(-np.pi, np.pi, 30)
COS, SIN = np.cos(X), np.sin(X)
plt.plot(X,COS)
plt.plot(X, SIN)
plt.show()
```

```python
X = np.linspace(0, np.pi, 30)
COS, SIN = np.cos(X), np.sin(X)
Y = np.linspace(0, np.pi, 30)
plt.loglog(X,1-COS,label = 'cosine') 
plt.loglog(X, X - SIN,label = 'sine')
plt.legend(loc='upper left')
plt.show()
```

```python
X = np.linspace(0, np.pi, 30)
COS, SIN = np.cos(X), np.sin(X)
Y = np.linspace(0, np.pi, 30)
plt.loglog(Y, 1- COS)
plt.loglog(Y, Y - SIN)
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 20)) # to get a big figure

theta = np.linspace(0,12*np.pi,10000) # lots of values of theta
r = np.exp(np.sin(theta)) - 2*np.cos(4*theta) + (np.sin((2*theta - np.pi)/24))**5 # corresponding values of r

plt.plot(r*np.cos(theta),r*np.sin(theta)) # x = r cos(thta), y = r sin(theta)
plt.axis('equal')
plt.show()
```
